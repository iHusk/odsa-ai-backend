### Example 6: Integer where a string is expected

`date` expects a string, but we pass an integer. Will Pydantic coerce it or reject it?

**Your prediction:** Pass or Fail?

### Playground Takeaway

**Pydantic catches shape problems, not meaning problems.**

- Wrong type? Caught.
- Missing required field? Caught.
- Nonsensical value in a valid type? Passes.
- Empty string or empty list? Passes.
- Extra fields? Ignored by default (Pydantic silently drops them).

This is why extraction quality depends on BOTH the schema AND the prompt. The schema enforces shape. The prompt guides meaning.

## 5. Extraction Deep Dive

Let's look at what is actually happening inside `extract_single()`. Claude's SDK uses **constrained decoding** via `output_format` — this guarantees valid JSON matching your schema at the token level. No manual parsing, no markdown stripping, no `json.loads()`.

### Behind the Scenes: How It Used to Work

Before `output_format`, developers had to serialize the schema into the prompt and parse JSON manually:

```python
# The OLD way (pre-2025) — don't use this
schema_json = json.dumps(schema_class.model_json_schema(), indent=2)
system_prompt = f"Return ONLY valid JSON matching this schema:\n{schema_json}"

response = client.messages.create(...)
raw_text = response.content[0].text

# Hope Claude didn't wrap it in markdown...
cleaned = strip_markdown_fences(raw_text)     # Could crash
parsed = json.loads(cleaned)                   # Could also crash
result = schema_class.model_validate(parsed)   # Could ALSO crash
```

Three parsing steps that could each fail. `output_format` eliminates all of them.

````python
# This is how extract_single() works under the hood:
# It uses messages.parse() with output_format to guarantee valid output

import anthropic

# The actual implementation in extractor.py:
print("""
def extract_single(document_text, schema_class):
    response = client.messages.parse(
        model="claude-sonnet-4-5",
        max_tokens=1024,
        temperature=0.0,
        system=SYSTEM_PROMPT,
        messages=[{
            "role": "user",
            "content": f"Extract structured data from this document:\\n\\n{document_text}"
        }],
        output_format=schema_class,  # <-- Pydantic class goes here
    )
    return response.parsed_output   # Already a typed Pydantic object
""")

print("Key insight: output_format tells Claude to use constrained decoding.")
print("The response is GUARANTEED to match your schema. No json.loads() needed.")
````

### What's Happening: Constrained Decoding

When you pass `output_format=MemoExtraction` to `messages.parse()`:

1. The SDK sends your Pydantic schema to Claude
2. Claude uses **constrained decoding** — at each token, it can only generate tokens that would produce valid JSON matching your schema
3. The response comes back as a `ParsedMessage` object
4. `.parsed_output` is already a fully typed Pydantic instance — `MemoExtraction`, not a dict

This is fundamentally different from asking Claude to "please return JSON" in the system prompt. Constrained decoding makes invalid output **impossible**, not just unlikely.

````python
# Live demo: extract_single returns a Pydantic object, not a dict
doc_path = _root / "data" / "northbrook" / "memo_security_update.md"
demo_text = doc_path.read_text()

demo_result = extract_single(demo_text, MemoExtraction)

# It's already a Pydantic object — no parsing needed
print(f"Type: {type(demo_result).__name__}")
print(f"Is MemoExtraction? {isinstance(demo_result, MemoExtraction)}")
print()

# Access fields directly as Python attributes
print(f"Author: {demo_result.author}")
print(f"Date:   {demo_result.date}")
print()

# Convert to dict when you need it
print("As dict:")
print(json.dumps(demo_result.model_dump(), indent=2))
````

## 6. Batch Processing

One document is easy. But real pipelines process many documents at once. The `extract_batch()` function:

- Loops over a list of file paths
- Calls `extract_single()` for each document
- Wraps each call in `try/except` so one bad document cannot crash the entire batch
- Returns a list of `ExtractionResult` objects (each knows whether it succeeded or failed)

Let's process a batch of Northbrook memos.

````python
# List the memo files we will process
data_dir = _root / "data" / "northbrook"

memo_files = sorted(str(p) for p in data_dir.glob("memo_*.md"))

print(f"Found {len(memo_files)} memo files:")
for f in memo_files:
    print(f"  {f}")
````

````python
# Run the batch -- this will make one API call per document
# Watch the output: [OK] for successes, errors for failures
results = extract_batch(memo_files, MemoExtraction)
````

````python
# Separate successes and failures
successes = [r for r in results if r.success]
failures = [r for r in results if not r.success]

print(f"\n=== Batch Summary ===")
print(f"Total documents: {len(results)}")
print(f"Successes:       {len(successes)}")
print(f"Failures:        {len(failures)}")
````

````python
# Inspect the successful extractions
print("=== Successful Extractions ===\n")
for r in successes:
    print(f"File: {Path(r.file_path).name}")
    print(f"  Author:  {r.data.author}")
    print(f"  Date:    {r.data.date}")
    print(f"  Subject: {r.data.subject}")
    print(f"  Points:  {len(r.data.key_points)} key points")
    print(f"  Actions: {len(r.data.action_items)} action items")
    print()
````

````python
# Inspect any failures -- what went wrong and why?
if failures:
    print("=== Failed Extractions ===\n")
    for r in failures:
        print(f"File:  {Path(r.file_path).name}")
        print(f"Error: {r.error[:200]}")
        print()
else:
    print("No failures -- all documents extracted successfully.")
````

**The key observation:** The batch finished. It did not crash. Even if some documents failed, you know exactly which ones and why. This is the batch processing pattern -- every document gets its own `try/except`.

### Exercise 1: Schema Mismatch

What happens when you extract a document with the **wrong** schema? Does it crash? Or does it silently produce bad data?

Let's find out. We will extract a **policy document** using `MemoExtraction` (wrong schema) and then `PolicyExtraction` (right schema).

````python
# Extract a policy document with the WRONG schema (MemoExtraction)
policy_text = (_root / "data" / "northbrook" / "remote_work_policy.md").read_text()

print("=== Extracting policy document with MemoExtraction (WRONG schema) ===\n")
try:
    wrong_result = extract_single(policy_text, MemoExtraction)
    print("Extraction succeeded -- but is the data correct?")
    print(wrong_result.model_dump_json(indent=2))
except Exception as e:
    print(f"Extraction failed: {e}")
````

````python
# Now extract the SAME document with the CORRECT schema (PolicyExtraction)
print("=== Extracting policy document with PolicyExtraction (CORRECT schema) ===\n")
correct_result = extract_single(policy_text, PolicyExtraction)
print(correct_result.model_dump_json(indent=2))
````

````python
# Compare the two schemas side by side -- what fields does each expect?
print("=== MemoExtraction expects ===")
for name, field_info in MemoExtraction.model_fields.items():
    print(f"  {name:20s}  {str(field_info.annotation):20s}  {field_info.description}")

print()
print("=== PolicyExtraction expects ===")
for name, field_info in PolicyExtraction.model_fields.items():
    print(f"  {name:20s}  {str(field_info.annotation):20s}  {field_info.description}")
````

### What happened?

The wrong schema **did not crash**. Pydantic said it was valid. But look at the data:

- `MemoExtraction` forced a policy document into memo-shaped fields. Claude filled in `author`, `subject`, and `key_points` because the schema demanded them -- but the values are **fabricated or forced** from content that does not have those concepts.
- `PolicyExtraction` asked for `policy_name`, `effective_date`, `sections`, `compliance_requirements` -- fields that actually match what the document contains.

**This is the danger of schema mismatch:** Pydantic validates shape, not meaning. The extraction "works" but the data is wrong. In a pipeline, this bad data flows downstream silently.
