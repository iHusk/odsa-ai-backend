## Step 1 — Install uv

### What is uv?

`uv` is an all-in-one tool that manages **both Python itself and your packages**. It replaces several tools you may have seen before (`pip`, `pyenv`, `venv`, `pipenv`) with one fast, unified command.

What uv handles for us:
- **Python version management** — installs and switches between Python versions (3.12, 3.13, etc.)
- **Package management** — installs libraries like `anthropic`, `numpy`, `pydantic`
- **Virtual environment creation** — sets up isolated project environments
- **Dependency resolution** — figures out which versions of packages work together

### Install it

Open a terminal and run **one** of the following:

**macOS / Linux:**
```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

**Windows (PowerShell):**
```powershell
powershell -ExecutionPolicy Bypass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

After installing, **close and reopen your terminal** (or in VS Code: `Cmd+Shift+P` → "Reload Window" on Mac, `Ctrl+Shift+P` on Windows).

### Verify it

In your terminal, run:
```bash
uv --version
```

You should see something like `uv 0.6.x`. If you get "command not found", try reopening your terminal.

## Step 3 — Set Up Your Virtual Environment

### What is a virtual environment?

A virtual environment is an **isolated container** for your project's Python and packages. Without one, every project on your machine shares the same packages — which means installing something for one class can break another.

The `.venv/` folder that we're about to create will contain:
- A copy of Python 3.12
- All the packages this course needs
- Nothing from your other projects

### Create it

Open a terminal in VS Code (`Terminal` → `New Terminal` from the menu bar) and run:

```bash
uv sync
```

This single command does everything:
1. Reads `pyproject.toml` (the project's recipe card — lists what Python version and packages we need)
2. Creates a `.venv/` folder
3. Installs Python 3.12 into it
4. Installs all required packages (`anthropic`, `pydantic`, `chromadb`, `numpy`, etc.)

You'll see a bunch of output as packages are downloaded and installed. If it finishes without errors, you're good.

### How the pieces fit together

```
┌─────────────────────────────────────────────────────┐
│                        uv                           │
│         (manages everything below)                  │
│                                                     │
│    ┌──────────────┐      ┌──────────────────────┐   │
│    │pyproject.toml│─────▶│      .venv/          │   │
│    │  (the recipe)│      │  ┌────────────────┐  │   │
│    │              │      │  │ Python 3.12    │  │   │
│    │ - anthropic  │      │  │ anthropic      │  │   │
│    │ - pydantic   │      │  │ pydantic       │  │   │
│    │ - chromadb   │      │  │ chromadb       │  │   │
│    │ - ...        │      │  │ ...            │  │   │
│    └──────────────┘      │  └────────────────┘  │   │
│                          │  (isolated packages) │   │
│                          └──────────────────────┘   │
│                                                     │
│         uv sync                                     │
│         reads the recipe → builds the environment   │
└─────────────────────────────────────────────────────┘
```

## Step 4 — Select Your Kernel

### What is a kernel?

A **kernel** is the engine that runs your notebook code. When you open a notebook in VS Code, you need to tell it *which Python* to use. We want it to use the one inside `.venv/` — not your system Python, not some other project's Python.

### How to select it

1. Look at the **top-right corner** of this notebook in VS Code — you'll see a kernel selector button
2. Click it
3. Choose **"Python Environments"**
4. Select the option that shows **`.venv`** (it should say something like `.venv (Python 3.12.x)`)

If you don't see `.venv` in the list:
- Make sure you ran `uv sync` in Step 3
- Try clicking the refresh icon in the kernel picker
- Try reloading VS Code (`Cmd+Shift+P` → "Reload Window")

### Now run the verification cell below

This is the moment of truth. If your kernel is set correctly, this cell will run and show PASS.

````python
import sys, os

# Check Python version
version = sys.version_info
if version.major == 3 and version.minor == 12:
    print(f'PASS: Python {version.major}.{version.minor}. RECORDED VERSIONS {version.micro}')
else:
    print(f'FAIL: Expected Python 3.12, got {version.major}.{version.minor}.{version.micro}')
    print('      Select the .venv kernel (top-right of this notebook)')

# Check we're running from .venv
in_venv = '.venv' in sys.executable or '.venv' in (os.environ.get('VIRTUAL_ENV', ''))
if in_venv:
    print(f'PASS: Running from virtual environment')
else:
    print(f'FAIL: Not running from .venv')
    print(f'      Current Python: {sys.executable}')
    print(f'      Select the .venv kernel (top-right of this notebook)')
````

If you see two **PASS** results above, your Python and kernel are configured correctly. If not, go back to Step 3 or Step 4 and follow the fix instructions.

## Step 5 — Verify Packages

Now let's make sure all the course packages were installed correctly by `uv sync`.

````python
# Verify all course packages are installed
import importlib

packages = [
    ('anthropic',    'Anthropic SDK'),
    ('pydantic',     'Pydantic'),
    ('chromadb',     'ChromaDB'),
    ('voyageai',     'Voyage AI'),
    ('dotenv',       'python-dotenv'),
    ('numpy',        'NumPy'),
]

all_ok = True
for mod, label in packages:
    try:
        importlib.import_module(mod)
        print(f'PASS: {label}')
    except ImportError:
        print(f'FAIL: {label} — not installed')
        all_ok = False

if all_ok:
    print('\nAll packages installed.')
else:
    print('\nSome packages missing. Try running: uv sync')
````

**Did you get a FAIL?** 

Sometimes after running `uv sync`, the kernel needs to be reloaded to see newly installed packages.

1. In VS Code: `Cmd+Shift+P` (Mac) or `Ctrl+Shift+P` (Windows)
2. Type **"Reload Window"** and press Enter
3. Re-run the cell above

**Important:** When you reload the window, all variables in the notebook are cleared. You'll need to re-run cells from the top. This is normal and expected — it's how notebooks work.

## Step 8 — Set Up Your Workspace

Throughout this course, you'll work from a **personal copy** of each session notebook. Here's why and how.

### The problem

If you edit notebooks directly in the `notebooks/` folder and the instructor pushes an update, `git pull` will create a conflict. Now you're stuck resolving merge conflicts instead of learning AI.

### The solution: `my_work/`

```
odsa-ai-backend/
├── notebooks/              ← Master copies (don't edit these)
│   ├── session_0_1_setup.ipynb
│   ├── session_1_1.ipynb
│   └── ...
│
├── my_work/                ← YOUR space (safe from updates)
│   ├── session_1_1.ipynb   ← Work here
│   └── ...
│
├── src/                    ← Shared modules (imports work from both folders)
└── data/                   ← Course dataset
```

The `my_work/` folder is **git-ignored** — it's invisible to git. The instructor can push updates all day long and your work is untouched.

### Your routine before each class

1. Open GitHub Desktop → click **"Fetch origin"** then **"Pull"**
2. Copy the new session notebook from `notebooks/` into `my_work/`
3. Open the copy from `my_work/` and work there

Run the cell below to create `my_work/` and copy your first notebook.

````python
import os, shutil
from pathlib import Path

# Find repo root by walking up until we find pyproject.toml
_here = Path(".").resolve()
_root = _here
while _root != _root.parent:
    if (_root / "pyproject.toml").exists():
        break
    _root = _root.parent
repo_root = _root

# Source: current session notebook
src_notebook = Path("session_1_1.ipynb").resolve()

# Destination: my_work folder at repo root
my_work = repo_root / "notebooks" / "my_work" / "1_1"
os.makedirs(my_work, exist_ok=True)

dst = my_work / "session_1_1.ipynb"

if src_notebook.exists():
    if not dst.exists():
        shutil.copy2(src_notebook, dst)
        print(f"✓ Copied to {dst.relative_to(repo_root)}")
    else:
        print(f"⚠ Already exists: {dst.relative_to(repo_root)}")
else:
    print(f"✗ Source not found: {src_notebook}")
````
